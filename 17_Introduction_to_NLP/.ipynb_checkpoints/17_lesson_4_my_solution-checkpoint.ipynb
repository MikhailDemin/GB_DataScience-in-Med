### Предобработка


Все модели, с которыми мы будем работать далее, работают с предобработанными текстами, поэтому сперва проведем стандартную предобработку текста:

```python
words_regex = re.compile('\w+')

def find_words(A, regex = words_regex):
    tokens =  regex.findall(A.lower())
    return [w for w in tokens if w.isalpha() and len(w) >= 3]


stopwords_list = stopwords.words('russian')

# mystem = Mystem()
# def lemmatize(words, lemmer = mystem, stopwords = stopwords_list):
#     lemmas = lemmer.lemmatize(' '.join(words))
#     return [w for w in lemmas if not w in stopwords 
#             and w.isalpha()]

def lemmatize(words, lemmer = morph, stopwords = stopwords_list):
    lemmas = [lemmer.parse(w)[0].normal_form for w in words]
    return [w for w in lemmas if not w in stopwords 
            and w.isalpha()]

def preprocess(A):
    return (lemmatize(find_words(A)))
```

```python
preprocessed_text = list(tqdm(map(preprocess, df['A']), total=len(df)))
```

```python
df['A'] = preprocessed_text
df.sample(3)
```

### Модель LDA


Первая модель, которую мы рассмотрим, LDA - латентное размещение Дирихле. Воспользуемся реализацией из библиотеки gensim.

```python
!pip install -U gensim
```

```python
!pip install --upgrade gensim
```

```python
from gensim.models import *
from gensim import corpora
```

Модель использует векторное представление документов, например, мешок слов (bow), поэтому сперва создадим словарь:

```python
dictionary = corpora.Dictionary(df['A'])

dictionary.filter_extremes(no_below = 10, no_above = 0.9, keep_n=None) # игнорируем слова, которые встречаются реже 10 раз или составляют более 0.9 словаря 
dictionary.save('rpp.dict')
```

Векторизуем документы:

```python
corpus = [dictionary.doc2bow(A) for A in df['A']]
corpora.MmCorpus.serialize('rpp.model', corpus) 
```

Теперь можем обучать модель:

```python
%time
lda = ldamodel.LdaModel(corpus, id2word=dictionary, num_topics=20, chunksize=50, update_every=1, passes=2)
```

Посмотрим на получившиеся темы:

```python
lda.show_topics(num_topics=10, num_words=10, formatted=True)
```

На полученные темы можно посмотреть, изобразив их на плоскости с помощью библиотеки __pyLDAvis__. Чтобы спроецировать темы на плоскость, внутри применяется PCA-разложение.

```python
!pip install pyLDAvis
```

```python
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis #import pyLDAvis.gensim as gensimvis

%time
vis_data = gensimvis.prepare(lda, corpus, dictionary)
pyLDAvis.display(vis_data)
```

Посмотрим на метрики качества получившейся модели: перплексию и среднюю когерентность:

```python
print(lda.log_perplexity(corpus))
```

```python
print('Персплексия: ', np.exp(lda.log_perplexity(corpus)))
```

```python
coherence_model_lda = CoherenceModel(model=lda, texts=df['A'], dictionary=dictionary, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('Средняя когерентность: ', coherence_lda)
```

Теперь можно подобрать оптимальное количество тем, опираясь на значение метрик:

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
topics_list = [5, 10, 15, 20, 25]
coherences = []

for num in tqdm(topics_list):
    lda = ldamodel.LdaModel(corpus, id2word=dictionary, num_topics=num, chunksize=50, update_every=1, passes=2)
    coherences.append(CoherenceModel(model=lda, texts=df['A'], dictionary=dictionary, coherence='c_v').get_coherence())

plt.plot(topics_list, coherences)
plt.xlabel("Число тем")
plt.ylabel("Средняя когерентность")
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
X_train.Content = X_train.Content.astype('U')
X_test.Content = X_test.Content.astype('U')
_, _, hist = plt.hist(X_train.Content.apply(lambda text: len(text.split())), bins='auto')
hist
```

Кроме этого, нужно перенумеровать как-то слова.

```python
from collections import Counter

words_counter = Counter((word for text in X_train.Content for word in text.lower().split()))

word2idx = {
    '': 0,
    '<unk>': 1
}
for word, count in words_counter.most_common():
    if count < 10:
        break
        
    word2idx[word] = len(word2idx)
    
print('Words count', len(word2idx))
```

**Задание** Сконвертируйте данные

```python
def convert(texts, word2idx, max_text_len):
    data = np.zeros((len(texts), max_text_len), dtype=np.int32)
    
    for inx, text in enumerate(texts):
        result = []
        for word in text.split():
            if word in word2idx:
                result.append(word2idx[word])
        padding = [0]*(max_text_len - len(result))
        data[inx] = np.array(padding + result[-max_text_len:], dtype=np.int32)
    return data
```

```python
X_train = convert(X_train.Content, word2idx, 75)
X_test = convert(X_test.Content, word2idx, 75)
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Embedding, GlobalMaxPooling1D, Dropout
```

```python
model = Sequential([
    Embedding(input_dim=len(word2idx), output_dim=64, input_shape=(X_train.shape[1],)),
    GlobalMaxPooling1D(),
    Dense(units=10, activation='relu'),
    Dense(units=10, activation='relu'),
    Dense(units=1, activation='sigmoid')
])

model.summary()
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size=128, epochs=7, 
          validation_data=(X_test, y_test))
```

```python
eval_loss, eval_acc = model.evaluate((X_test))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import skimage.color as cs
import scipy.signal as sig
import imageio
```

```python
img = imageio.v2.imread('15_1_dragonfly_head.jpg')
img = img.astype(np.float32) / 255.
print(img.shape)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(img)
```

```python
# создадим функция изменяющую цветовую схему и насыщенность
def image_process(img):
    pic = img.copy()
    pic = cs.rgb2hsv(pic)
    pic[:, :, 0] *= 7
    img_new = cs.hsv2rgb(pic)
    return img_new
```

```python
image = image_process(img)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(image)
```

## –ó–∞–¥–∞—á–∞ 3-1. 
–î–∞–Ω—ã –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞—Ä–ø–ª–∞—Ç –∏–∑ –≤—ã–±–æ—Ä–∫–∏ –≤—ã–ø—É—Å–∫–Ω–∏–∫–æ–≤: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. –ü–æ—Å—á–∏—Ç–∞—Ç—å (–∂–µ–ª–∞—Ç–µ–ª—å–Ω–æ –±–µ–∑ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏—Ö –º–µ—Ç–æ–¥–æ–≤ –Ω–∞–ø–æ–¥–æ–±–∏–µ std, var, mean) —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ, —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ, —Å–º–µ—â–µ–Ω–Ω—É—é –∏ –Ω–µ—Å–º–µ—â–µ–Ω–Ω—É—é –æ—Ü–µ–Ω–∫–∏ –¥–∏—Å–ø–µ—Ä—Å–∏–π –¥–ª—è –¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏.
–ú–æ–∂–Ω–æ –∑–∞—Ç–µ–º –ø–æ—Å—á–∏—Ç–∞—Ç—å —Ç–µ –∂–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π —Å —Ü–µ–ª—å—é —Å–∞–º–æ–ø—Ä–æ–≤–µ—Ä–∫–∏.

### –†–µ—à–µ–Ω–∏–µ:
[1] C—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ –∑–∞—Ä–ø–ª–∞—Ç –≤—ã–ø—É—Å–∫–Ω–∏–∫–æ–≤: $M(X)={\frac{1}{n}}{\displaystyle\sum_{i=1}^{n} x_i}$

[2] –°—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –∑–∞—Ä–ø–ª–∞—Ç –≤—ã–ø—É—Å–∫–Ω–∏–∫–æ–≤: $\sigma = \sqrt{\frac{\displaystyle\sum\limits_{i=1}^{n} (x_i - \overline{x})^2}{n}}$

[3] –°–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –∑–∞—Ä–ø–ª–∞—Ç –≤—ã–ø—É—Å–∫–Ω–∏–∫–æ–≤: $\sigma^2_{biased} = \frac{\displaystyle\sum\limits_{i=1}^{n} (x_i - \overline{x})^2}{n}$

[4] –ù–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –∑–∞—Ä–ø–ª–∞—Ç –≤—ã–ø—É—Å–∫–Ω–∏–∫–æ–≤: $\sigma^2_{unbiased} = \frac{\displaystyle\sum\limits_{i=1}^{n} (x_i - \overline{x})^2}{n - 1}$

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
def mean_salary(salary):
    if len(salary) > 0:
        mean_salary_1 = salary.sum() / len(salary)
        mean_salary_2 = salary.mean()
        print(f'–°—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ —Ä–∞–≤–Ω–æ {mean_salary_1} (–ø—Ä–æ–≤–µ—Ä–∫–∞ {mean_salary_2})')

def std_salary(salary):
    if len(salary) > 0:
        std_salary_1 = np.sqrt(((salary - salary.mean()) ** 2).sum() / len(salary))
        std_salary_2 = salary.std(ddof=0)
        print(f'–°—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ {std_salary_1} (–ø—Ä–æ–≤–µ—Ä–∫–∞ {std_salary_2})')

def variance_salary1(salary):
    if len(salary) > 0:
        variance_salary_1 = ((salary - salary.mean()) ** 2).sum() / len(salary)
        variance_salary_2 = salary.var(ddof=0)
        print(f'C–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏  {variance_salary_1} (–ø—Ä–æ–≤–µ—Ä–∫–∞ {variance_salary_2})')

def variance_salary2(salary):
    if len(salary) > 0:
        variance_salary2_1 = ((salary - salary.mean()) ** 2).sum() / (len(salary) - 1)
        variance_salary2_2 = salary.var(ddof=1)
        print(f'–ù–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏  {variance_salary2_1} (–ø—Ä–æ–≤–µ—Ä–∫–∞ {variance_salary2_2})')
```

```python
salary = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
mean_salary(salary)
std_salary(salary)
variance_salary1(salary)
variance_salary2(salary)
```

## –ó–∞–¥–∞—á–∞ 3-2.

### –†–µ—à–µ–Ω–∏–µ:
[1] –í–µ—Ä–æ—è—Ç–Ω–æ—Å—å –≤—ã–Ω—É—Ç—å 0 –±–µ–ª—ã—Ö –º—è—á–µ–π –∏–∑ 1-–≥–æ —è—â–∏–∫–∞ –∏ 3 –±–µ–ª—ã—Ö –∏–∑ 2-–≥–æ —è—â–∏–∫–∞: $P_1=\left({\frac{{C_{5}^0}*{C_{3}^2}}{C_{8}^2}}\right)*\left({\frac{{C_{5}^3}*{C_{7}^1}}{C_{12}^4}}\right)=\left({\frac{{\frac{5!}{0!(5-0)!}}*{\frac{3!}{2!(3-2)!}}}{\frac{8!}{2!(8-2)!}}}\right)*\left({\frac{{\frac{5!}{3!(5-3)!}}*{\frac{7!}{1!(7-1)!}}}{\frac{12!}{4!(12-4)!}}}\right)$

[2] –í–µ—Ä–æ—è—Ç–Ω–æ—Å—å –≤—ã–Ω—É—Ç—å 1 –±–µ–ª—ã–π –º—è—á –∏–∑ 1-–≥–æ —è—â–∏–∫–∞ –∏ 2 –±–µ–ª—ã—Ö –∏–∑ 2-–≥–æ —è—â–∏–∫–∞: $P_2=\left({\frac{{C_{5}^1}*{C_{3}^1}}{C_{8}^2}}\right)*\left({\frac{{C_{5}^2}*{C_{7}^2}}{C_{12}^4}}\right)=\left({\frac{{\frac{5!}{1!(5-1)!}}*{\frac{3!}{1!(3-1)!}}}{\frac{8!}{2!(8-2)!}}}\right)*\left({\frac{{\frac{5!}{2!(5-2)!}}*{\frac{7!}{2!(7-2)!}}}{\frac{12!}{4!(12-4)!}}}\right)$

[3] –í–µ—Ä–æ—è—Ç–Ω–æ—Å—å –≤—ã–Ω—É—Ç—å 2 –±–µ–ª—ã—Ö –º—è—á–∞ –∏–∑ 1-–≥–æ —è—â–∏–∫–∞ –∏ 1 –±–µ–ª—ã–π –º—è—á –∏–∑ 2-–≥–æ —è—â–∏–∫–∞: $P_3=\left({\frac{{C_{5}^2}*{C_{3}^0}}{C_{8}^2}}\right)*\left({\frac{{C_{5}^1}*{C_{7}^3}}{C_{12}^4}}\right)=\left({\frac{{\frac{5!}{2!(5-2)!}}*{\frac{3!}{0!(3-0)!}}}{\frac{8!}{2!(8-2)!}}}\right)*\left({\frac{{\frac{5!}{1!(5-1)!}}*{\frac{7!}{3!(7-3)!}}}{\frac{12!}{4!(12-4)!}}}\right)$

[4] –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ 3 –º—è—á–∞ –±–µ–ª—ã–µ: $P=P_1+P_2+P_3$

```python
from scipy.special import comb
p1 = ((comb(5, 0, exact = False, repetition=False) * (comb(3, 2, exact = False, repetition=False))) / comb(8, 2, exact = False, repetition=False)) * (((comb(5, 3, exact = False, repetition=False)) * (comb(7, 1, exact = False, repetition=False))) / comb(12, 4, exact = False, repetition=False))
p2 = ((comb(5, 1, exact = False, repetition=False) * (comb(3, 1, exact = False, repetition=False))) / comb(8, 2, exact = False, repetition=False)) * (((comb(5, 2, exact = False, repetition=False)) * (comb(7, 2, exact = False, repetition=False))) / comb(12, 4, exact = False, repetition=False))
p3 = ((comb(5, 2, exact = False, repetition=False) * (comb(3, 0, exact = False, repetition=False))) / comb(8, 2, exact = False, repetition=False)) * (((comb(5, 1, exact = False, repetition=False)) * (comb(7, 3, exact = False, repetition=False))) / comb(12, 4, exact = False, repetition=False))
print(f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—å –≤—ã–Ω—É—Ç—å 0 –±–µ–ª—ã—Ö –º—è—á–µ–π –∏–∑ 1-–≥–æ —è—â–∏–∫–∞ –∏ 3 –±–µ–ª—ã—Ö –∏–∑ 2-–≥–æ —è—â–∏–∫–∞ = {round(p1, 7)}\n'
      f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—å –≤—ã–Ω—É—Ç—å 1 –±–µ–ª—ã–π –º—è—á –∏–∑ 1-–≥–æ —è—â–∏–∫–∞ –∏ 2 –±–µ–ª—ã—Ö –∏–∑ 2-–≥–æ —è—â–∏–∫–∞ = {round(p2, 7)}\n'
      f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—å –≤—ã–Ω—É—Ç—å 2 –±–µ–ª—ã—Ö –º—è—á–∞ –∏–∑ 1-–≥–æ —è—â–∏–∫–∞ –∏ 1 –±–µ–ª—ã–π –∏–∑ 2-–≥–æ —è—â–∏–∫–∞ = {round(p3, 7)}\n'
      f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Ä–æ–≤–Ω–æ –¥–≤–∞ –º—è—á–∞ –±–µ–ª—ã–µ = {round((p1 + p2 + p3), 7)}')
```

## –ó–∞–¥–∞—á–∞ 3-3.

### –†–µ—à–µ–Ω–∏–µ:
#### –§–æ—Ä–º—É–ª–∞ –ë–∞–π–µ—Å–∞
–ß—Ç–æ–±—ã –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Å–æ–±—ã—Ç–∏—è $B$ –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏, —á—Ç–æ —Å–æ–±—ã—Ç–∏–µ $A$ —É–∂–µ –ø—Ä–æ–∏–∑–æ—à–ª–æ, –∏—Å–ø–æ–ª—å–∑—É—é—Ç —Ñ–æ—Ä–º—É–ª—É –ë–∞–π–µ—Å–∞:
$$P(B\:|\:A) = \frac{P(B) \cdot P(A\:|\:B)}{P(A)}$$
–í –¥–∞–Ω–Ω–æ–π —Ñ–æ—Ä–º—É–ª–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å $P(B\;|\;A)$ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –∞–ø–æ—Å—Ç–µ—Ä–∏–æ—Ä–Ω–æ–π, —Ç–æ –µ—Å—Ç—å –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–π –ø–æ—Å–ª–µ —Ç–æ–≥–æ, –∫–∞–∫ –ø—Ä–æ–∏–∑–æ—à–ª–æ —Å–æ–±—ã—Ç–∏–µ $A$. 

–ê –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å $P(B)$ ‚Äî –∞–ø—Ä–∏–æ—Ä–Ω–æ–π, –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–π –¥–æ –∏—Å–ø—ã—Ç–∞–Ω–∏—è.
–ü—Ä–µ–∂–¥–µ —á–µ–º —É–∑–Ω–∞–ª–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤—ã—Å—Ç—Ä–µ–ª–∞, –º—ã —Ä–∞—Å–ø–æ–ª–∞–≥–∞–ª–∏ —Ç–æ–ª—å–∫–æ –∞–ø—Ä–∏–æ—Ä–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é —Ç–æ–≥–æ, —á—Ç–æ –≤—ã—Å—Ç—Ä–µ–ª –±—ã–ª –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω, –∏ –æ–Ω–∞ —Ä–∞–≤–Ω—è–ª–∞—Å—å 1/3 ‚Äî –µ—Å–ª–∏ —É—á–µ—Å—Ç—å, —á—Ç–æ —É –≤—Å–µ—Ö —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–≤ –±—ã–ª–∏ —Ä–∞–≤–Ω—ã–µ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏ –≤—ã—Å—Ç—Ä–µ–ª–∏—Ç—å.

–ó–Ω–∞—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤—ã—Å—Ç—Ä–µ–ª–∞, –º—ã –º–æ–∂–µ–º –∑–∞–Ω–æ–≤–æ –æ—Ü–µ–Ω–∏—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–∞ ‚Äî –≤–µ–¥—å —Ç–æ—á–Ω–æ—Å—Ç—å —É –Ω–∏—Ö —Ä–∞–∑–Ω–∞—è. –≠—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –±—É–¥—É—Ç –∞–ø–æ—Å—Ç–µ—Ä–∏–æ—Ä–Ω—ã–º–∏, —Ç–∞–∫ –∫–∞–∫ –æ–Ω–∏ –ø–æ–ª—É—á–µ–Ω—ã –ø–æ—Å–ª–µ –ø–æ—Å—Ç—É–ø–ª–µ–Ω–∏—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ —Ç–æ–º, —á—Ç–æ –≤—ã—Å—Ç—Ä–µ–ª –±—ã–ª —É–¥–∞—á–Ω—ã–º.

–°–æ–±—ã—Ç–∏–µ–º $A$ –±—É–¥–µ–º —Å—á–∏—Ç–∞—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏–µ –≤ –º–∏—à–µ–Ω—å, –∞ —Å–æ–±—ã—Ç–∏—è–º–∏ $B_1, B_2$ –∏ $B_3$ ‚Äî —á—Ç–æ –≤—ã—Å—Ç—Ä–µ–ª —Å–æ–≤–µ—Ä—à–∏–ª –ø–µ—Ä–≤—ã–π, –≤—Ç–æ—Ä–æ–π –∏–ª–∏ —Ç—Ä–µ—Ç–∏–π —Å–ø–æ—Ä—Ç—Å–º–µ–Ω:

$P(A\:|\:B_1) = 0.9$

$P(A\:|\:B_2) = 0.8$

$P(A\:|\:B_3) = 0.6$

a). –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—Ä–µ–ª—è–ª –ø–µ—Ä–≤—ã–π —Å–ø–æ—Ä—Ç—Å–º–µ–Ω ‚Äî –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏, —á—Ç–æ –≤—ã—Å—Ç—Ä–µ–ª –±—ã–ª —É–¥–∞—á–Ω—ã–º, ‚Äî –±—É–¥–µ–º –Ω–∞—Ö–æ–¥–∏—Ç—å –ø–æ —Ñ–æ—Ä–º—É–ª–µ:
$$P(B_1\:|\:A) = \frac{P(B_1) \cdot P(A\:|\:B_1)}{P(A)}=\frac{\frac{1}{3} \cdot 0.9}{\frac{1}{3} \cdot 0.9 + \frac{1}{3} \cdot 0.8 + \frac{1}{3} \cdot 0.6}$$

–±). –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—Ä–µ–ª—è–ª –≤—Ç–æ—Ä–æ–π —Å–ø–æ—Ä—Ç—Å–º–µ–Ω ‚Äî –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏, —á—Ç–æ –≤—ã—Å—Ç—Ä–µ–ª –±—ã–ª —É–¥–∞—á–Ω—ã–º, ‚Äî –±—É–¥–µ–º –Ω–∞—Ö–æ–¥–∏—Ç—å –ø–æ —Ñ–æ—Ä–º—É–ª–µ:
$$P(B_2\:|\:A) = \frac{P(B_2) \cdot P(A\:|\:B_2)}{P(A)}=\frac{\frac{1}{3} \cdot 0.8}{\frac{1}{3} \cdot 0.9 + \frac{1}{3} \cdot 0.8 + \frac{1}{3} \cdot 0.6}$$

–≤). –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—Ä–µ–ª—è–ª —Ç—Ä–µ—Ç–∏–π —Å–ø–æ—Ä—Ç—Å–º–µ–Ω ‚Äî –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏, —á—Ç–æ –≤—ã—Å—Ç—Ä–µ–ª –±—ã–ª —É–¥–∞—á–Ω—ã–º, ‚Äî –±—É–¥–µ–º –Ω–∞—Ö–æ–¥–∏—Ç—å –ø–æ —Ñ–æ—Ä–º—É–ª–µ:
$$P(B_3\:|\:A) = \frac{P(B_3) \cdot P(A\:|\:B_3)}{P(A)}=\frac{\frac{1}{3} \cdot 0.6}{\frac{1}{3} \cdot 0.9 + \frac{1}{3} \cdot 0.8 + \frac{1}{3} \cdot 0.6}$$

```python
p_1 = 0.9
p_2 = 0.8
p_3 = 0.6
p_A = 1/3 * p_1 + 1/3 * p_2 + 1/3 * p_3
p_B1 = (1/3 * p_1) / p_A
p_B2 = (1/3 * p_2) / p_A
p_B3 = (1/3 * p_3) / p_A
print(f'–ü–æ–ª–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è –≤ –º–∏—à–µ–Ω—å –≤—Å–µ–º–∏ —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–∞–º–∏: ùëÉ(ùê¥) = {round(p_A, 7)}\n'
      f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—Ä–µ–ª—è–ª –ø–µ—Ä–≤—ã–π —Å–ø–æ—Ä—Ç—Å–º–µ–Ω: ùëÉ(ùêµ1|ùê¥) = {round(p_B1, 7)}\n'
      f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—Ä–µ–ª—è–ª –≤—Ç–æ—Ä–æ–π —Å–ø–æ—Ä—Ç—Å–º–µ–Ω: ùëÉ(ùêµ2|ùê¥) = {round(p_B2, 7)}\n'
      f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—Ä–µ–ª—è–ª —Ç—Ä–µ—Ç–∏–π —Å–ø–æ—Ä—Ç—Å–º–µ–Ω: ùëÉ(ùêµ3|ùê¥) = {round(p_B3, 7)}\n'
      f'–ü—Ä–æ–≤–µ—Ä–∫–∞ –≤–µ—Ä–Ω–æ—Å—Ç–∏ —Ä–∞—Å—á–µ—Ç–æ–≤: ùëÉ(ùêµ1|ùê¥) + ùëÉ(ùêµ2|ùê¥) + ùëÉ(ùêµ3|ùê¥) = {round((p_B1 + p_B2 + p_B3), 7)}')
```

#### –û—Ç–≤–µ—Ç 3-3: –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –≤—ã—Å—Ç—Ä–µ–ª –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω: 
#### a). –ø–µ—Ä–≤—ã–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º $P(B_1\:|\:A) = 39,13\%$
#### –±). –≤—Ç–æ—Ä—ã–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º $P(B_2\:|\:A) = 34,78\%$
#### –≤). —Ç—Ä–µ—Ç—å–∏–º —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–º $P(B_3\:|\:A) = 26,09\%$

## –ó–∞–¥–∞—á–∞ 3-4.
–í —É–Ω–∏–≤–µ—Ä—Å–∏—Ç–µ—Ç –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç—ã A –∏ B –ø–æ—Å—Ç—É–ø–∏–ª–æ —Ä–∞–≤–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤, –∞ –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç C —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ –ø–æ—Å—Ç—É–ø–∏–ª–æ —Å—Ç–æ–ª—å–∫–æ –∂–µ, —Å–∫–æ–ª—å–∫–æ –Ω–∞ A –∏ B –≤–º–µ—Å—Ç–µ. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–∞ A —Å–¥–∞—Å—Ç –ø–µ—Ä–≤—É—é —Å–µ—Å—Å–∏—é, —Ä–∞–≤–Ω–∞ 0.8. –î–ª—è —Å—Ç—É–¥–µ–Ω—Ç–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–∞ B —ç—Ç–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ä–∞–≤–Ω–∞ 0.7, –∞ –¥–ª—è —Å—Ç—É–¥–µ–Ω—Ç–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–∞ C - 0.9. –°—Ç—É–¥–µ–Ω—Ç —Å–¥–∞–ª –ø–µ—Ä–≤—É—é —Å–µ—Å—Å–∏—é.
–ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –æ–Ω —É—á–∏—Ç—Å—è: a). –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ A –±). –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ B –≤). –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ C?

### –†–µ—à–µ–Ω–∏–µ:
#### –§–æ—Ä–º—É–ª–∞ –ë–∞–π–µ—Å–∞
–ß—Ç–æ–±—ã –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Å–æ–±—ã—Ç–∏—è $B$ –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏, —á—Ç–æ —Å–æ–±—ã—Ç–∏–µ $A$ —É–∂–µ –ø—Ä–æ–∏–∑–æ—à–ª–æ, –∏—Å–ø–æ–ª—å–∑—É—é—Ç —Ñ–æ—Ä–º—É–ª—É –ë–∞–π–µ—Å–∞:
$$P(B\:|\:A) = \frac{P(B) \cdot P(A\:|\:B)}{P(A)}$$
[1] A –∏ B —Ä–∞–≤–Ω—ã –∏ —è–≤–ª—è—é—Ç—Å—è –ø–æ–ª–æ–≤–∏–Ω–æ–π C, —Ç–æ –æ—á–µ–≤–∏–¥–Ω–æ, —á—Ç–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Å–ª—É—á–∞–π–Ω–æ–≥–æ –≤—ã–±–æ—Ä–∞:

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É—á–∏—Ç—Å—è –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ –ê: $P(F_A) = 0.25$

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É—á–∏—Ç—Å—è –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ B: $P(F_B) = 0.25$

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É—á–∏—Ç—Å—è –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ C: $P(F_C) = 0.5$

[2] –í–æ–∑—å–º–µ–º –∏–∑ —É—Å–ª–æ–≤–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ —Å–¥–∞—á–∏ –∫–∞–∂–¥–æ–≥–æ —Å—Ç—É–¥–µ–Ω—Ç–∞:

$P(F_A\:|\:A) = 0.8$

$P(F_B\:|\:A) = 0.7$

$P(F_C\:|\:A) = 0.9$

[3] –ù–∞–π–¥–µ–º –ø–æ–ª–Ω—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å–¥–∞—Ç—É—Ç –≤—Å–µ —Å—Ç—É–¥–µ–Ω—Ç—ã:

$P(A) = P(F_A) \cdot P(F_A\:|\:A) + P(F_B) \cdot P(F_B\:|\:A) + P(F_C) \cdot P(F_C\:|\:A) $

[4] –ù–∞–π–¥–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ —Å–æ–≥–ª–∞—Å–Ω–æ —Ñ–æ—Ä–º—É–ª–µ –ë–∞–π–µ—Å–∞:

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É—á–∏—Ç—Å—è –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ A: $ùëÉ(F_A) = \frac{P(F_A) \cdot P(F_A\:|\:A)}{P(A)}$

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É—á–∏—Ç—Å—è –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ B: $ùëÉ(F_B) = \frac{P(F_B) \cdot P(F_B\:|\:A)}{P(A)}$

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É—á–∏—Ç—Å—è –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ C: $ùëÉ(F_C) = \frac{P(F_C) \cdot P(F_C\:|\:A)}{P(A)}$

```python
p_1 = 0.8
p_2 = 0.7
p_3 = 0.9
p_A = (0.25 * p_1) + (0.25 * p_2) + (0.5 * p_3)
p_F_A = (0.25 * p_1) / p_A
p_F_B = (0.25 * p_2) / p_A
p_F_C = (0.5 * p_3) / p_A
print(f'–ü–æ–ª–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å–¥–∞—Ç—É—Ç –≤—Å–µ —Å—Ç—É–¥–µ–Ω—Ç—ã: ùëÉ(ùê¥) = {round(p_A, 7)}\n'
      f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É—á–∏—Ç—Å—è –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ A: ùëÉ(ùêπùê¥) = {round(p_F_A, 7)}\n'
      f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É—á–∏—Ç—Å—è –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ B: ùëÉ(ùêπùêµ) = {round(p_F_B, 7)}\n'
      f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É—á–∏—Ç—Å—è –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ C: ùëÉ(ùêπùê∂) = {round(p_F_C, 7)}\n'
      f'–ü—Ä–æ–≤–µ—Ä–∫–∞ –≤–µ—Ä–Ω–æ—Å—Ç–∏ —Ä–∞—Å—á–µ—Ç–æ–≤: ùëÉ(FA) + ùëÉ(Fùêµ) + ùëÉ(FC) = {round((p_F_A + p_F_B + p_F_C), 7)}')
```

#### –û—Ç–≤–µ—Ç 3-4: –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É—á–∏—Ç—Å—è: 
#### a). –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ A: $ùëÉ(ùêπ_ùê¥) = 24,24(24)\%$
#### –±). –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ B: $P(F_B) = 21,21(21)\%$
#### –≤). –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ C: $P(F_C) = 54,54(54)\%$

## –ó–∞–¥–∞—á–∞ 3-5.
–£—Å—Ç—Ä–æ–π—Å—Ç–≤–æ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ —Ç—Ä–µ—Ö –¥–µ—Ç–∞–ª–µ–π. –î–ª—è –ø–µ—Ä–≤–æ–π –¥–µ—Ç–∞–ª–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã–π—Ç–∏ –∏–∑ —Å—Ç—Ä–æ—è –≤ –ø–µ—Ä–≤—ã–π –º–µ—Å—è—Ü —Ä–∞–≤–Ω–∞ 0.1, –¥–ª—è –≤—Ç–æ—Ä–æ–π - 0.2, –¥–ª—è —Ç—Ä–µ—Ç—å–µ–π - 0.25. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –≤ –ø–µ—Ä–≤—ã–π –º–µ—Å—è—Ü –≤—ã–π–¥—É—Ç –∏–∑ —Å—Ç—Ä–æ—è:
–∞). –≤—Å–µ –¥–µ—Ç–∞–ª–∏
–±). —Ç–æ–ª—å–∫–æ –¥–≤–µ –¥–µ—Ç–∞–ª–∏
–≤). —Ö–æ—Ç—è –±—ã –æ–¥–Ω–∞ –¥–µ—Ç–∞–ª—å
–≥). –æ—Ç –æ–¥–Ω–æ–π –¥–æ –¥–≤—É—Ö –¥–µ—Ç–∞–ª–µ–π?

### –†–µ—à–µ–Ω–∏–µ:

```python
p_1 = 0.1
p_2 = 0.2
p_3 = 0.25
p_all = p_1 * p_2 * p_3
p_only_2 = ((1-p_1)*p_2*p_3)+(p_1*(1-p_2)*p_3)+(p_1*p_2*(1-p_3))
p_at_least_1 = 1-((1-p_1)*(1-p_2)*(1-p_3))
print(f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –∏–∑ —Å—Ç—Ä–æ—è –≤—ã–π–¥—É—Ç –≤—Å–µ –¥–µ—Ç–∞–ª–∏: ùëÉ(all) = {p_all}\n'
      f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –∏–∑ —Å—Ç—Ä–æ—è –≤—ã–π–¥—É—Ç —Ç–æ–ª—å–∫–æ –¥–≤–µ –¥–µ—Ç–∞–ª–∏: ùëÉ(only 2) = {p_only_2}\n'
      f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, –∏–∑ —Å—Ç—Ä–æ—è –≤—ã–π–¥–µ—Ç —Ö–æ—Ç—è –±—ã –æ–¥–Ω–∞ –¥–µ—Ç–∞–ª—å: ùëÉ(at least 1) = {p_at_least_1}\n'
      f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, –∏–∑ —Å—Ç—Ä–æ—è –≤—ã–π–¥—É—Ç –æ—Ç –æ–¥–Ω–æ–π –¥–æ –¥–≤—É—Ö –¥–µ—Ç–∞–ª–µ–π: ùëÉ(1 till 2) = {p_at_least_1 - p_all}\n')
```

#### –û—Ç–≤–µ—Ç 3-5: –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –≤ –ø–µ—Ä–≤—ã–π –º–µ—Å—è—Ü –≤—ã–π–¥—É—Ç –∏–∑ —Å—Ç—Ä–æ—è: 
#### a). –≤—Å–µ –¥–µ—Ç–∞–ª–∏: $ùëÉ_{all} = 0,5\%$
#### –±). —Ç–æ–ª—å–∫–æ –¥–≤–µ –¥–µ—Ç–∞–ª–∏: $P_{only~2} = 8\%$
#### –≤). —Ö–æ—Ç—è –±—ã –æ–¥–Ω–∞ –¥–µ—Ç–∞–ª—å: $P_{at~least~1} = 46\%$
#### –≥). –æ—Ç –æ–¥–Ω–æ–π –¥–æ –¥–≤—É—Ö –¥–µ—Ç–∞–ª–µ–π: $P_{1~till~2} = 45,5\%$

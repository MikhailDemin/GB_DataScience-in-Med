```r
set.seed(4)
samp<-rnorm(50,7,3)
samp
```

```r
z<-qnorm(0.975)
z
```

```r
SE<-3/sqrt(50)
SE
```

```r
lolv<-mean(samp)-z*SE
uplv<-mean(samp)+z*SE
CI<-c(lolv, uplv)
CI
```

```r
7>= CI[1] & 7 <=CI[2]
```

```r
bigpar(1,3)
set.seed(3)
plot(7 +c(-4,4), c(1,1), type = "n", xlab = "mu", ylab = "доверительные интервалы", ylim = c(1,100), main = "N=10")
abline(v = 7, col = "brown", lwd=2)
for (i in 1:100) {
    sam <- rnorm(10,7,3)
    SE <- sd(sam)/sqrt(10)
    CI <- c(mean(sam)-z*SE, mean(sam)+z*SE)
    catch <- 7>= CI[1] & 7 <=CI[2]
    color <- ifelse(catch, "blue", "red")
    lines(CI, c(i,i), col = color, lwd = 2)
}

set.seed(3)

plot(7 +c(-3,3), c(1,1), type = "n", xlab = "mu", ylab = "доверительные интервалы", ylim = c(1,100), main = "N=50")
abline(v = 7, col = "brown", lwd=2)
for (i in 1:100) {
    sam <- rnorm(50,7,3)
    SE <- sd(sam)/sqrt(50)
    CI <- c(mean(sam)-z*SE, mean(sam)+z*SE)
    catch <- 7>= CI[1] & 7 <=CI[2]
    color <- ifelse(catch, "blue", "red")
    lines(CI, c(i,i), col = color, lwd = 2)
}

set.seed(3)

plot(7 +c(-3,3), c(1,1), type = "n", xlab = "mu", ylab = "доверительные интервалы", ylim = c(1,100), main = "N=100")
abline(v = 7, col = "brown", lwd=2)
for (i in 1:100) {
    sam <- rnorm(100,7,3)
    SE <- sd(sam)/sqrt(100)
    CI <- c(mean(sam)-z*SE, mean(sam)+z*SE)
    catch <- 7>= CI[1] & 7 <=CI[2]
    color <- ifelse(catch, "blue", "red")
    lines(CI, c(i,i), col = color, lwd = 2)
}
```

```r
dat<-read.csv("18_cardio_train.csv",sep = ";")
head(dat)
```

```r
dat$age/365
trunc(dat$age/365)
```

```r
dat <-dat %>% mutate(age_years=(trunc(age/365)))
head(dat)
```

```r
tidy_set <- dat %>% filter((ap_lo<200&ap_lo>20)&(ap_hi<300&ap_hi>40))
head(tidy_set)
dim(tidy_set)
```

```r
head(tidy_set[tidy_set$ap_hi<tidy_set$ap_lo,])
.tidy_set <- tidy_set[tidy_set$ap_hi>tidy_set$ap_lo,]
dim(.tidy_set)
dim(tidy_set)
```

```r
.women<-.tidy_set$ap_lo[.tidy_set$gender==1]
.men<-.tidy_set$ap_lo[.tidy_set$gender==2]
length(.women)
length(.men)
```

```r
mypar(1,2)
qqnorm(.women, main = 'женщины')
qqline(.women)

qqnorm(.men, main = 'мужчины')
qqline(.men)

# рассчитаем характеристики описательной статистики
infer<- .tidy_set %>% group_by(gender) %>% summarise(
    mu = mean(ap_lo),
    k = qt(0.975, length(ap_lo)-1),
    se = sd(ap_lo)/sqrt(length(ap_lo)),
    lo_level = mean(ap_lo) - k*se,
    hi_level = mean(ap_lo) + k*se)
infer
```

```r
#CI for women
ci_w <- c(infer[1,5], infer[1,6])
ci_w
```

```r
ci_w <- as.numeric(ci_w)
ci_w
```

```r
ci_m <- as.numeric(c(infer[2,5],infer[2,6]))
ci_m
```

```r
mypar(1,1)
plot(mean(.women), col = 2, lwd = 3, xlim = c(0.5,2.5), ylim = c(80,83),
    ylab = 'среднее диастолическое давление',
    main = 'интервальная оценка среднего диастолического давления для мужчин и женщин')
interval_w = c(80.75,80.92)
lines(x = c(1,1), y = interval_w, col = 'red', lwd = 3)
points(1.5, mean(.men), col = 3, lwd = 2)
interval_m = c(82.05,82.29)
lines(x = c(1.5,1.5), y = interval_m, col = 'blue', lwd = 3)
legend('topleft', c('women', 'men'), fill = c('red', 'blue'))
```

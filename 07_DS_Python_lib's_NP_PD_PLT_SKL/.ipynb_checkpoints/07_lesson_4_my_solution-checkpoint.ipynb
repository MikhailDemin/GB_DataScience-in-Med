__8.__ Создать одномерный массив а, состоящий из последовательности целых чисел от 0 до 12. Поменять форму этого массива,
чтобы получилась матрица ```A``` (двумерный массив Numpy), состоящая из 4 строк и 3 столбцов. Получить матрицу ```At``` путем 
транспонирования матрицы ```A```. Получить матрицу ```B```, умножив матрицу ```A``` на матрицу ```At``` с помощью матричного умножения. Какой размер
имеет матрица ```B```? Получится ли вычислить обратную матрицу для матрицы ```B``` и почему?

__Решение:__<br>

````python
# Создать одномерный массив а, состоящий из последовательности целых чисел от 0 до 12.
a = np.arange(0, 12)
a
````

````python
# Поменять форму этого массива, чтобы получилась матрица A (двумерный массив Numpy), состоящая из 4 строк и 3 столбцов.
A = a.reshape(4, 3)
A
````

````python
# Получить матрицу At путем транспонирования матрицы A.
At = A.T
At
````

````python
# Получить матрицу B, умножив матрицу A на матрицу At с помощью матричного умножения.
B = A.dot(At)
B
````

````python
# Какой размер имеет матрица B?
B.shape
````

````python
# Получится ли вычислить обратную матрицу для матрицы B и почему?
B_inv=np.linalg.inv(B)
B_inv
````

__Ответ:__ Вычислить обратную матрицу для матрицы B не получится, т.к. это вырожденная (сингулярная) матрица.

__12.__ Поменяйте его форму так, чтобы получилась квадратная матрица ```C```. Получите матрицу ```D```, поэлементно прибавив матрицу ```B``` из
предыдущего вопроса к матрице ```C```, умноженной на 10. Вычислите определитель, ранг и обратную матрицу ```D_inv``` для ```D```.

__Решение:__<br>

````python
# Поменяйте его форму так, чтобы получилась квадратная матрица C.
C = c.reshape(4, 4)
C
````

````python
# Получите матрицу D, поэлементно прибавив матрицу B из предыдущего вопроса к матрице C, умноженной на 10.
print(f'Матрица B:\n{B}')
C_10 = C * 10
print(f'Матрица C*10:\n{C_10}')
D = C_10 + B
print(f'Матрица D = C*10 + B:\n{D}')
````

````python
# Вычислите определитель, ранг и обратную матрицу D_inv для D.
print(f"Определитель матрицы D равен:  {np.linalg.det(D)}")
print(f'Ранг матрицы D равен: {np.linalg.matrix_rank(D)}')
D_inv=np.linalg.inv(D)
print(f'Обратная матрица D_inv:\n{D_inv}')
````

__13.__ Приравняйте к нулю отрицательные числа в матрице ```D_inv```, а положительные - к единице. Убедитесь, что в матрице ```D_inv``` остались только нули и единицы. С помощью функции ```numpy.where```, используя матрицу ```D_inv``` в качестве маски, а матрицы ```B``` и ```C``` - в качестве источников данных, получите матрицу ```E``` размером 4x4. Элементы матрицы ```E```, для которых соответствующий элемент матрицы ```D_inv``` равен 1, должны быть равны соответствующему элементу матрицы ```B```, а элементы матрицы ```E```, для которых соответствующий элемент матрицы ```D_inv``` равен 0, должны быть равны соответствующему элементу матрицы ```C```.

__Решение:__<br>

````python
# Приравняйте к нулю отрицательные числа в матрице D_inv, а положительные - к единице.
# Убедитесь, что в матрице D_inv остались только нули и единицы.
D_inv = np.where(D_inv<0, 0, 1)
D_inv
````

````python
# С помощью функции numpy.where, используя матрицу D_inv в качестве маски, а матрицы B и C - в качестве источников данных,
# получите матрицу E размером 4x4. Элементы матрицы E, для которых соответствующий элемент матрицы D_inv равен 1, должны быть
# равны соответствующему элементу матрицы B, а элементы матрицы E, для которых соответствующий элемент матрицы D_inv равен 0,
# должны быть равны соответствующему элементу матрицы C.
print(f'Матрица B:\n{B}')
print(f'Матрица C:\n{C}')
print(f'Матрица D_inv:\n{D_inv}')
E = np.where(D_inv == 1, B, C)
print(f'Матрица E:\n{E}')
````
